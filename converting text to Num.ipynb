```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy
```

```python
corpus = ["Machine learning is super fun",
"Python is super, super cool",
"Statistics is cool, too",
"Data science is fun",
"Python is great for machine learning",
"I like football",
"Football is great to watch"]

vectorizer = CountVectorizer(stop_words = 'english')

dtm = vectorizer.fit_transform(corpus)

pd.DataFrame(dtm.toarray(),index=corpus,columns=vectorizer.get_feature_names())
```

```python
corpus = ["Machine learning is super fun",
"Python is super, super cool",
"Statistics is cool, too",
"Data science is fun",
"Python is great for machine learning",
"I like football",
"Football is great to watch"]
vectorizer = CountVectorizer(stop_words='english', ngram_range=(1,2), max_features=5)
dtm = vectorizer.fit_transform(corpus)
pd.DataFrame(dtm.toarray(),index=corpus,columns=vectorizer.get_feature_names()).head(5)
```
